```python
aelist = ['QaE', 'QbE', 'QcE', 'QdE', 'QeE', 'QfE', 'QgE', 'QhE', 'QiE', 'QjE', 'QkE', 'QlE', 'QmE', 'QnE', 'QoE', 'QpE', 'QqE', 'QrE', 'QsE', 'QtE']
```

```python
# 1분위 값과 3분위값 확인
for i in aelist:
    print(i ,'25%', ':', train[i].quantile(.25))
    print(i ,'75%', ':', train[i].quantile(.75))
    print('*'*30)
```

```python
# 1분위값 이하는 Small, 3분위 이상은 Big, 나머지는 Middle
for i in aelist:
    train[i+'_new'] = np.where(train[i]<=train[i].quantile(.25),'Small',train[i])
    train[i+'_new'] = np.where(train[i]>=train[i].quantile(.75),'Big',train[i+'_new'])
    train[i+'_new'] = np.where((train[i]>=train[i].quantile(.25)) & (train[i]<=train[i].quantile(.75)),'Middle',train[i+'_new'])
   
```

tp_에서 서로 반대되는 항목들의 차이가 가장 큰 값을 대표값으로 설정

```python
# 어떤 항목이 차이가 가장 큰지 구하기
for i in range(0, len(train)):
    tp_list=[]
    first = abs(tps.iloc[i,0]-tps.iloc[i,5])
    second = abs(tps.iloc[i,1]-tps.iloc[i,6])
    third = abs(tps.iloc[i,2]-tps.iloc[i,7])
    forth = abs(tps.iloc[i,3]-tps.iloc[i,8])
    fifth = abs(tps.iloc[i,4]-tps.iloc[i,9])
    
    tp_list = [first, second, third, forth, fifth]
    tps.loc[i, 'max_num'] = max(tp_list)
        
    if tps.loc[i, 'max_num'] == tp_list[0]:
        tps.loc[i,'max_ele'] = 'first'
    elif tps.loc[i, 'max_num'] == tp_list[1]:
        tps.loc[i,'max_ele'] = 'second'
    elif tps.loc[i, 'max_num'] == tp_list[2]:
        tps.loc[i,'max_ele'] = 'third'
    elif tps.loc[i, 'max_num'] == tp_list[3]:
        tps.loc[i,'max_ele'] = 'forth'
    else :
        tps.loc[i,'max_ele'] = 'fifth'
```

```python
# 차이가 큰 항목을 기준으로 어떤 값이 대표값인지 설정
tps['tp'] = 0
for i in range(0, len(train)):
    if tps.loc[i,'max_ele'] == 'first':
        if tps.iloc[i,0]-tps.iloc[i,5] >0:
            tps.loc[i,'tp'] = 'tp01'
        else:
            tps.loc[i,'tp'] = 'tp06'
            
    elif tps.loc[i,'max_ele'] == 'second':
        if tps.iloc[i,1]-tps.iloc[i,6] >0:
            tps.loc[i,'tp'] = 'tp02'
        else:
            tps.loc[i,'tp'] = 'tp07'

    elif tps.loc[i,'max_ele'] == 'third':
        if tps.iloc[i,2]-tps.iloc[i,7] >0:
            tps.loc[i,'tp'] = 'tp03'
        else:
            tps.loc[i,'tp'] = 'tp08'

    elif tps.loc[i,'max_ele'] == 'forth':
        if tps.iloc[i,3]-tps.iloc[i,8] >0:
            tps.loc[i,'tp'] = 'tp04'
        else:
            tps.loc[i,'tp'] = 'tp09'

    else:
        if tps.iloc[i,4]-tps.iloc[i,9] >0:
            tps.loc[i,'tp'] = 'tp05'
        else:
            tps.loc[i,'tp'] = 'tp10'
```

```python
aes = train.loc[:, 'QaE_new':'QtE_new']
aes.head()
```

```python
for i in range(0, len(train)):
    s = 0
    m = 0
    b = 0
    for j in aelist:
        if aes.loc[i,j] == 'Small':
            s +=1
        elif aes.loc[i,j] == 'Middle':
            m +=1
        else : 
            b += 1
            
    if s == m :
        aes.loc[i,'ae'] = 'sm'
        continue
    elif m == b:
        aes.loc[i,'ae'] = 'mb'
        continue
    elif b ==s :
        aes.loc[i,'ae'] = 'bs'
        continue
        
    aenum = [s,m,b]
    aenum = sorted(aenum)
    if aenum[2] == s :
        aes.loc[i,'ae'] = 's'
    elif aenum[2] == m :
        aes.loc[i,'ae'] = 'm'
    else:
        aes.loc[i,'ae'] = 'b'
    
    
```
